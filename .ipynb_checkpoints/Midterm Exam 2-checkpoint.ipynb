```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
import pandas as pd
import matplotlib
matplotlib.use('MacOSX')
%matplotlib inline
import matplotlib.pyplot as plt
X = pd.read_csv ("/users/pisonwaters/desktop/midterm_train.csv")
y = X.pop("y")
```

```python
X.describe()
```

```python
# Impute NAs with mean
X=X.fillna(value=0)

# Confirm the code is correct
X.describe()
```

```python
# Drop the variables I don't feel like dealing with for this tutorial
X.drop(["x24", "x29", "x30", "x32", "x37"], axis=1, inplace=True)
```

```python

model = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=42)
model.fit(X, y)
```

```python
model.oob_score_
```

```python
y_oob = model.oob_prediction_
print "c-stat: ", roc_auc_score(y, y_oob)
```

```python
model.feature_importances_
```

```python
%%timeit
model = RandomForestRegressor(1000, oob_score=True, n_jobs=1, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=1000, 
                              oob_score=True, 
                              n_jobs=-1, 
                              random_state=42, 
                              max_features="auto", 
                              min_samples_leaf=5)
model.fit(X, y)
roc = roc_auc_score(y, model.oob_prediction_)
print "C-stat: ", roc
```
