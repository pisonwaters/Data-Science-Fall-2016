```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cross_validation import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score
#from sklearn.feature_extraction.text import TfidfTransformer
#from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
#df= pd.read_csv("desktop/Naive Bayes Data.csv",sep='\t', names=['Indicator', 'Text'])
```

```python
df= pd.read_csv ("desktop/Naive Data.txt",sep='\t', names=['Indicator', 'Text'])
```

```python
df.head()
```

```python
df=df.replace('ham',0)

#df_Indicator = pd.get_dummies()
  
```

```python
df=df.replace('spam',1)
```

```python
df
```

```python
#TFIDF Vectorizer, just like before
stopset = set(stopwords.words('english'))
vectorizer = TfidfVectorizer(use_idf=True, lowercase=True, strip_accents='ascii', stop_words=stopset)
```

```python
#in this case our dependent variable will be ham as 0 (good email) or 1 (spam)
y = df.Indicator
```

```python
X = vectorizer.fit_transform(df.Text)
```

```python
#5572 observations x 2 unique words.
print y.shape
print X.shape
```

```python
#Test Train Split as usual
X_train, X_test,y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#we will train a naive_bayes classifier
clf = naive_bayes.MultinomialNB()
clf.fit(X_train, y_train)
```

```python
#we can test our model's accuracy like this:

roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
```

```python
sms_reviews_array=np.array(["I see the letter B on my car"])

sms_review_vector = vectorizer.transform(sms_reviews_array)

print clf.predict(sms_review_vector)
```
